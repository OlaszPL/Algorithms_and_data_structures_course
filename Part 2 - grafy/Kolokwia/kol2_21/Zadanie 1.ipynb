### Omówienie algorytmu

Najpierw, przy pomocy algorytmu Floyda-Warshalla, wyznaczam długości wszystkich ścieżek między
każdą parą wierzchołków w grafie. Następnie, korzystając z obliczonych dystansów, tworzę nowy graf,
który jest reprezentowany poprzez 2-wymiarową tablicę list sąsiedztwa. $ G[u][v] $ oznacza, że z danego położenia (Carol znajduje się na pozycji $ u $, a Max na pozycji $ v $), możliwe jest przejście do takiej konfiguracji, ($ u_2 $, $ v_2 $), że Carol przechodzi z wierzchołka $ u $ na $ u2 $, a Max z $ v $ na $ v2 $. Ponieważ mogą albo oboje się nadal poruszać, albo jedna z osób może stać w miejscu, możliwa jest sytuacja, w której $ u_2 $ = $ u $ lub $ v_2 $ = $ v $, ale nigdy nie zajdzie jednocześnie $ u_2 $ = $ u $ i $ v_2 $ = $ v $, bo wtedy oboje by stali w miescu i algorytm by się zapętlił. Odrzucamy również przypadek, w którym obie osoby zamieniają się miejscami, tzn. $ u_2 = v $ i $ v_2 = u $, co wynika z założeń zadania. Przy tworzeniu takiego grafu, rozważam wszystkie możliwe nowe położenia obu osób (lub to, że zostają w miejscu) i dla każdej takiej konfiguracji sprawdzam, czy odległość dzieląca obie osoby po zmianie pozycji na nową jest większa lub równa $ d $. Jeżeli możliwe jest przejście na nowe położenie obu osób, dodaję krawedź (krawędzią jest para wierzchołków docelowych po danym przejściu obu osób) do nowego grafu w reprezentacji listowej.
<br><br>
Po utworzeniu grafu, wyznaczam zwykłym algorytmem DFS ścieżkę do celu, czyli kolejne kroki, jakie muszą wykonać obie osoby. Ponieważ ta ścieżka nie musi być najkrótsza, wyznaczam dowolną sekwencę ruchów, która prowadzi nas do rozwiązania. Za każdym razem zapamiętuję, skąd przyszedłem, w celu odtworzenia rozwiązania.
<br><br>
Ponieważ jest powiedziane, że rozwiązanie istnieje zawsze, po zakończeniu algorytmu DFS, tworzę listę kolejnch kroków - rozwiązanie.

### Złożoność

###### Obliczeniowa:
$ O(V^4) $ - czas potrzebny na zbudowanie nowego grafu (dominujący element)
###### Pamięciowa:
$ O(V^4) $ <br>
Wynika to stąd, że dla każdego usytowania Carol ($ V $ możliwych bieżących wierzchołków), mamy potencjalnie $ O(V) $ możliwości wyboru następnego wierzchołka. Przy każdym takim wyborze, Maks może stać na jednym z $ O(V) $ wierzchołków i wybrać również jeden z $ O(V) $ nowych wierzchołków. Taka złożoność jest osiągana jedynie w przypadku grafu pełnego oraz takiej odległości $ d $, dla której obie osoby mogą dotrzeć do około $ \frac{V}{2} $ wierzchołków w każdym kroku.

### Implementacja

```python
def floyd_warshall(G: 'graph represented by adjacency matrix'):
    n = len(G)
    inf = float('inf')
    W = [[inf] * n for _ in range(n)]

    for i in range(n):
        for j in range(n):
            if G[i][j]:
                W[i][j] = G[i][j]
            elif i == j:
                W[i][j] = 0

    for t in range(n):
        for i in range(n):
            for j in range(n):
                if W[i][t] + W[t][j] < W[i][j]:
                    W[i][j] = W[i][t] + W[t][j]

    return W


def create_graph(M, dist, d):
    n = len(M)
    inf = float('inf')
    G = [[[] for _ in range(n)] for _ in range(n)]

    for x1 in range(n):
        for x2 in range(n):
            if x1 != x2 and not M[x1][x2]: continue
            for y1 in range(n):
                for y2 in range(n):
                    if y1 != y2 and not M[y1][y2]: continue
                    if (x1 == x2 and y1 == y2) or (x2 == y1 and y2 == x1): continue
                    if d <= dist[x2][y2] < inf:
                        G[x1][y1].append((x2, y2))

    return G


def get_path(parents, coords):
    path = []

    while coords:
        path.append(coords)
        u, v = coords
        coords = parents[u][v]

    path.reverse()
    return path


def keep_distance(M, x, y, d):
    dist = floyd_warshall(M)
    G = create_graph(M, dist, d)
    n = len(G)

    visited = [[False] * n for _ in range(n)]
    parents = [[None] * n for _ in range(n)]

    def dfs(u, v):
        visited[u][v] = True
        if u == y and v == x:
            return True
        for u2, v2 in G[u][v]:
            if not visited[u2][v2]:
                parents[u2][v2] = u, v
                if dfs(u2, v2):
                    return True
        return False

    dfs(x, y)

    return get_path(parents, (y, x))
```

# Sposób #2 (wzorcowy)

### Omówienie algorytmu

Niewiele się różni od mojego pomysłu. Jedyna różnica dotyczy tego, że nie tworzymy fizycznie nowego grafu, ale na bieżąco sprawdzamy, do jakiej konfiguracji możemy przejść i zapamiętujemy odpowiednio w $ 2 $-wymiarowej tablicy $ visited $ konfiguracje, które już sprawdziliśmy.

### Złożoność

###### Obliczeniowa:
$ O(V^4) $
###### Pamięciowa:
$ O(V^2) $ - bo już nie tworzymy nowego grafu, a jedynie spamiętujemy odwiedzone pary wierzchołków (konfiguracje) w tablicy o wymiarach $ V \times V $.

### Implementacja

```python
def floyd_warshall(G: 'graph represented by adjacency matrix'):
    n = len(G)
    inf = float('inf')
    W = [[inf] * n for _ in range(n)]

    for i in range(n):
        for j in range(n):
            if G[i][j]:
                W[i][j] = G[i][j]
            elif i == j:
                W[i][j] = 0

    for t in range(n):
        for i in range(n):
            for j in range(n):
                if W[i][t] + W[t][j] < W[i][j]:
                    W[i][j] = W[i][t] + W[t][j]

    return W


def get_path(parents, coords):
    path = []

    while coords:
        path.append(coords)
        u, v = coords
        coords = parents[u][v]

    path.reverse()
    return path


def keep_distance(M, x, y, d):
    dist = floyd_warshall(M)
    G = create_graph(M, dist, d)
    n = len(G)
    inf = float('inf')

    visited = [[False] * n for _ in range(n)]
    parents = [[None] * n for _ in range(n)]

    def dfs(u, v):
        visited[u][v] = True
        if u == y and v == x:
            return True
        
        for u2 in range(n):
            if u2 != u and not M[u][u2]: continue
            for v2 in range(n):
                if visited[u2][v2] or \
                    (v2 != v and not M[v][v2]) or \
                    (u == u2 and v == v2) or \
                    (u2 == v and v2 == u): 
                    continue
                if d <= dist[u2][v2] < inf:
                    parents[u2][v2] = u, v
                    dfs(u2, v2)
        return False

    dfs(x, y)

    return get_path(parents, (y, x))
```
